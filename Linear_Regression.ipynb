```python
+import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
data = {'YearsExperience':[1,2,3,4,5,6,7,8,9],'Salary':[40000,45000,50000,55000,60000,65000,70000,75000,80000]}
df=pd.DataFrame(data)
# Define features and target
A = df[['YearsExperience']]
y = df['Salary']

# Split the data into training and test sets
A_train, A_test, y_train, y_test = train_test_split(A, y, test_size=0.2, random_state=42)

# Initialize and train the model
model = LinearRegression()
model.fit(A_train, y_train)

# Predict on the test set
y_pred=model.predict(A_test)
y_experience = float(input("Enter years of Experience:"))
y_predict=model.predict([[y_experience]])
print("Predicted Salary:",y_predict[0])

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
